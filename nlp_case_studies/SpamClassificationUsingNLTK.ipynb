```python
import pandas as pd
messages= pd.read_csv("SMSSpamCollection",sep='\t',names=["label","message"])
```

```python
messages
```

```python
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
#corpus
```

Create Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
X= cv.fit_transform(corpus).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=10000,formatter=dict(float=lambda x: "%.3g" % x))
#X
```

```python
#cv.vocabulary_
```

```python
##Output Features

y= pd.get_dummies(messages["label"])
```

```python
y=y.iloc[:,0].values
```

```python
##Train test split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model= MultinomialNB().fit(X_train,y_train)
```

```python
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(ngram_range=(1, 2),max_features=2500)
X= tfv.fit_transform(corpus).toarray()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_tfidf_model = MultinomialNB().fit(X_train,y_train)
```

```python
y_pred= spam_tfidf_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
