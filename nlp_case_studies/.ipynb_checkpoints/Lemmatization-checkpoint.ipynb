```python
##WordNet Lemmatizer
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
'''
POS - Noun-n
verb-v
adjective-a
adverb-r
'''

lemmatizer.lemmatize("going",pos='v')
```

```python
lemmatizer.lemmatize("going",pos='r')
```

```python
words = ["running","eaten","eating", "playing", "studies", "easily", "connected"]
```

```python
for word in words:
    print(word+"--->"+lemmatizer.lemmatize(word,pos='v'))
```

```python
for word in words:
    print(word+"--->"+lemmatizer.lemmatize(word,pos='r'))
```

```python
lemmatizer.lemmatize("fairly",pos='a'),lemmatizer.lemmatize("sportingly",pos='v')
```
