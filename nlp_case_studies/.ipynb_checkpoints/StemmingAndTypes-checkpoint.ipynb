```python
#Stemming
#Classification Problem 
#Comments of product is a positive or negative review.
##Reviews-->Eating,Eat,eaten.//[going,gone,goes-->go]
##Eat is the root word.

words = ["running","eaten","eating", "playing", "studies", "easily", "connected"]
```

```python
##PorterStemmer
from nltk.stem import PorterStemmer
stemming = PorterStemmer()
```

```python
for word in words:
    print(word+"---->"+stemming.stem(word))
```

```python
stemming.stem('congratulations')
```

```python
stemming.stem("sitting")
```

```python
##RegexpStemmer Class
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer= RegexpStemmer('ing$|s$|able$',min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

```python
##SnowballStemmer
from nltk.stem import SnowballStemmer
```

```python
sbs=SnowballStemmer('english')
```

```python
for word in words:
    print(word+"--->"+sbs.stem(word))
```

```python
##PorterStemmer
stemming.stem("fairly"),stemming.stem("sportingly")
```

```python
##SnowballStemmer
sbs.stem("fairly"),sbs.stem("sportingly")
```
