```python
text = """Dr. A. P. J. Abdul Kalam was born in Rameswaram, Tamil Nadu, India.
He served as the President of India from 2002 to 2007.
Kalam worked at ISRO and DRDO and played a major role in India's missile program.
He received the Bharat Ratna in 1997.
"""
```

```python
import nltk
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
words=nltk.word_tokenize(text)
```

```python
tag_ele = nltk.pos_tag(words)
```

```python
nltk.ne_chunk(tag_ele).draw()
```
