```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification

X,y = make_classification(
    n_samples=1000,
    n_features =3,
    n_redundant=1,
    n_classes=2,
    random_state=99
)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5,algorithm ='auto',p=2)
#algorithm = auto it choose best algorithm
#the value of p=2 is by default it means Euclidean Distance,
#p=1 means manhatton distance
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(accuracy_score(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
param_grid = {
    'n_neighbors': [1,2,3,4],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}
```

```python
from sklearn.model_selection import GridSearchCV
grid =GridSearchCV(estimator=knn,param_grid=param_grid,refit=True,cv=3)
```

```python
grid.fit(X_train,y_train)
y_pred = grid.predict(X_test)
```

```python
print(accuracy_score(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
```
