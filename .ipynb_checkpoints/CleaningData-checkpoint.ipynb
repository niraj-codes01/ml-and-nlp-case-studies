```python
#Cleaning Data
#handle missing values.
import pandas as pd
df = pd.read_csv("raw_data.csv")

df.isnull()
df.isna()

df.isnull().sum()
df.dropna() #Row drop
df.dropna(axis=1) #Column drop

df.fillna(0)
df.fillna(100)

cleaned_data = df.copy()
age_mean = cleaned_data["age"].mean()
cleaned_data["age"] = cleaned_data["age"].fillna(age_mean)

df.ffill() #Forward Fill

df.bfill() #Backward Fill

```

```python
#Handle the duplicate

df.duplicated()
df[["country","gender"]].duplicated()
df2 = df.copy()
df.drop_duplicates(inplace= True)
```

```python
df = df2
```

```python
df
```

```python
#Data Types

df.dtypes
df2= df.copy()

df2 = df2.fillna(0)

df2["age"] = df2["age"].astype("int64").copy()
df2.dtypes
```

```python
#Feature Engineering

df2 = df.copy()
#apply()
df2["tax"]  = df2["income"].apply(lambda x : "20%" if x>=60000 else "10%")

#map()
gender_map = {"Male" : "M", "Female": "F", "Unknown": "U"}
df2["gender"] = df2["gender"].map(gender_map)

#assign()
df2.assign(new_income = df2["income"] * 1.1)

#replace(old,new)
df2["country"] = df2["country"].replace("USA","US")

df2.rename(columns={"income":"Salary"})
df2.rename(index={1:"First"})

df2.sort_values("income")
sorted_df = df2.sort_values(["income","age"])
sorted_df.sort_index()

#Reset
sorted_df.reset_index()
sorted_df.reset_index(drop = True)

#Ranking

sorted_df["Ranking"] = sorted_df["income"].rank(ascending= False,method="dense")
```

```python
df2 = df.copy()

new_col_order = [col for col in df2.columns if col != "id"]+["id"]
print(new_col_order)
df2[new_col_order]
```

```python
#writing to CSV
df2 = df.copy()

df2 = df2.drop_duplicates()
df2 = df2.fillna(0)
df2 = df2.sort_values("income")

df2.reset_index(drop=True)

df2.to_csv("Sorted_data.csv")
```
